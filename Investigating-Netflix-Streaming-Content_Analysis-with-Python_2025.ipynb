```python
# 1] HOW TO APPROACH THE PROJECT "DATASET" 

# TASK 001 : COLLECT & CHECK THE LIST OF ALL MOVIES RELEASED IN THE 1990s : 

# ROW-WISE FILTERING/SUBSETTING TO RETURN ALL MOVIES RELEASED IN THE 1990s :

Movies_90s = netflix_df[(netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] <= 1999)]
Movies_90s
```

```python
print(f"Total number of Movies released in the 1990s: {len(Movies_90s)}") 
```

```python
# TASK 002 :Find the most frequent movie duration from the column "duration" in our 90s Movies: 
# OR, IN OTHER WORDS, WE MAY CALCULATE THE MODE OF THE GIVEN FIELD/COLUMN TO GET THE MOST FREQ. OCCURING VALUE (Using Python's Mode(). ) 

Mode_duration90s = Movies_90s['duration'].mode()[0]
print(f"So, the most frequent movie duration : {Mode_duration90s} minutes")
```

```python

# To "Count" the Frequency of each Duration ie ; the number of times a specific "Duration" has occured in our 90s Movies, we pass the value_counts() method to 
# the 'duration' column (which is basically considered as a Categorical-Data Column, here ) 

Count_duration90s = Movies_90s["duration"].value_counts()
Count_duration90s 
 
```
